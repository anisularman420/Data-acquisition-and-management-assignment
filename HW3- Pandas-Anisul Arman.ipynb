```python
import pandas as pd

# Add daily miles for 50 days
daily_miles = [
    55, 65, 75, 80, 75, 60, 90, 85, 78, 65,
    80, 75, 80, 95, 75, 80, 72, 68, 90, 85,
    75, 70, 80, 69, 87, 66, 80, 85, 75, 92,
    65, 75, 85, 70, 80, 70, 70, 85, 90, 65,
    75, 90, 75, 85, 75, 68, 85, 75, 90, 80
]

# Compute cumulative miles using a for loop
cumulative_miles = []
total = 0
for miles in daily_miles:
    total += miles
    cumulative_miles.append(total)


# Create a Pandas Series
cumulative_series = pd.Series(cumulative_miles)

# Show the cummulative Series
# Create a DataFrame
df = pd.DataFrame({
    "Day": range(1, 51),
    "Daily Total Miles": daily_miles,
    "Cumulative Miles": cumulative_miles
})

ride_df = df.set_index("Day")
```
