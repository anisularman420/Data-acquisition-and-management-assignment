```python
# import all the neccesary libraries
import pandas as pd
```

```python
# load the dataset  
df = pd.read_csv("airline.csv")  
```

```python
# calculate total flights  
df["total_flights"] = df["ontime"] + df["delayed"]  
```

```python

# calculate delay percentage  
df["delay_percentage"] = (df["delayed"] / df["total_flights"]) * 100  
```

```python
# total flights per airline  
airline_flights = df.groupby("airline")["total_flights"].sum()  
print(airline_flights) 
```

```python

# plot total flights per airline  
airline_flights.plot(kind="bar", title="total flights per airline")  
```

```python
# plot  total delayed per airline 
total_delayed_flights = df.groupby("airline")["delayed"].sum()  
total_delayed_flights.plot(kind="bar", title="total delayed per airline", color="orange") 
```

```python
# plot  total delayed per airline 
total_ontime_flights = df.groupby("airline")["ontime"].sum()  
total_ontime_flights.plot(kind="bar", title="total airline that are ontime ", color="orange") 
```

```python
# average delay percentage per airline  
airline_delay = df.groupby("airline")["delay_percentage"].mean()  
airline_delay
```

```python
# plot average delay percentage per airline  
airline_delay.plot(kind="bar", title="average delay percentage per airline", color="red")  
```

```python
# city with highest delay percentage  
city_delay = df.groupby("city")["delay_percentage"].mean()  
city_delay
```

```python
# total flights per city  
city_flights = df.groupby("city")["total_flights"].sum()  
print(city_flights)  
```

```python
# plot total flights per city  
city_flights.plot(kind="bar", title="total flights per city", color="green")  
```

```python
# plot average delay percentage per city  
city_delay.plot(kind="bar", title="average delay percentage per city", color="orange")  
```
